```python
import pandas as pd

games = pd.read_csv('games_tiancheng (1).csv')
games.head(10)
```

```python
mean = games.drop_duplicates(subset = 'Attempt', keep = 'first')['diff'].mean()
std = games.drop_duplicates(subset = 'Attempt', keep = 'first')['diff'].std()
```

```python
print('The mean difference of our strategy compared to the perfect foresight strategy across 778 iterations is: ', mean, 'percent')
print('The standard deviation difference of our strategy compared to the perfect foresight strategy across 778 iterations is: ', mean, 'percent')
```
